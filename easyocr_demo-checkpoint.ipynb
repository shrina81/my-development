```python
from PIL import Image
import easyocr

import cv2
from matplotlib import pyplot as plt
import numpy as np
reader = easyocr.Reader(['en'])
```

```python
img = cv2.imread("partnum3.png")
img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
```

```python
#run false prediction
result = reader.readtext(img)
print(result)
```

```python
result[0][0][0][0],result[0][0][2][1]
```

```python
img_pil = Image.open("partnum3.png")
img2 = img_pil.crop((result[0][0][0][0],result[0][0][0][1], result[0][0][2][0],result[0][0][2][1])) #top-left,right-bottom
```

```python
plt.imshow(img2)
```

```python
img = np.array(img2)
```

```python
plt.imshow(img)
```

```python
grayImage = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
thresh, bw = cv2.threshold(grayImage, 127, 255, cv2.THRESH_TRUNC)
```

```python
plt.imshow(bw,cmap='gray')
```

```python
dn_bw = cv2.fastNlMeansDenoising(bw,None,10,7,7)
```

```python
plt.imshow(dn_bw,cmap='gray')
```
