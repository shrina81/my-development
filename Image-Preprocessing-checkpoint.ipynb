```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import imutils
import easyocr
```

```python
#img = cv2.imread('The Real One.png')
#img = cv2.imread('clientCropped.jpg')
img = cv2.imread('5.jpg')
#img = cv2.imread('newcliientCropped.jpg')
#img = cv2.imread('testClientImage.jpg')
#img = cv2.imread('finalCropped.jpg')
#img = cv2.imread('shrinaCropped.jpg')
#img = cv2.imread('crop6.jpg')
#img_cropped = img[1350:1700,2500:3020]
#img_cropped = img[1300:1700,2500:3040]
#img_cropped = img[1300:1700,2500:3070]
#img_cropped = img[1300:1700,2500:3050]
img_cropped = img[1300:1700,2410:3070]
#(87, 83)
#(633, 367)

#img = cv2.imread('finalCropped1.jpg')
#img = cv2.imread('onlineCapture.jpg')
print(img_cropped.shape)

gray = cv2.cvtColor(img_cropped, cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.cvtColor(gray, cv2.COLOR_BGR2RGB))
```

```python
#crop image
w=img.shape[1]/2
h=img.shape[0]/2
center = tuple()
center = img.shape
x = center[1]/3
y = center[0]/3
# x,y=200,250

img = img[int(y):int(h), int(x):int(x+w)]
print(img.shape)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
ret, threshold = cv2.threshold(gray, 127, 255 , cv2.THRESH_TRUNC)
plt.imshow(cv2.cvtColor(threshold, cv2.COLOR_BGR2RGB))
```

```python
reader = easyocr.Reader(['en'])
result = reader.readtext(threshold)
result
```

```python
top_left = tuple(result[0][0][0])
print(top_left)
bottom_right = tuple(result[0][0][2])
print(bottom_right)
text = result[0][1]
print(text)
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python


img = cv2.rectangle(img_cropped,top_left,bottom_right,(0,255,0),4)
img = cv2.putText(img,text,top_left,font,4,(0,0,255),4,cv2.LINE_AA)
#plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.imshow(img)
plt.show()

```
